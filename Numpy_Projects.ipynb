**Expense Tracker**

```python
#Create an array of 28 random expenses between $10 and $200
import numpy as np
expenses = np.random.randint(10,200,28)
print("Expenses:",expenses)
#Calculate total of monthly expenses using np.sum() function
MaxExpense = np.sum(expenses)
print("Total expense for the month:",MaxExpense)
#Average daily expense
AverageExpense = MaxExpense/28
print("Average expense per day:",AverageExpense)
#print day with maxm expense
print("Day with max expense:",np.argmax(expenses))
#Total expense per week (split into 4 weeks, hint: reshape or slice)
expenses_per_week = expenses.reshape(4,7)
print("Expenses per week:",expenses_per_week)
```

**Correlation Tool**

```python
# Sequential arrays
import numpy as np
array1 = np.arange(1,21)
array2 = np.arange(21,41)
print("Array 1:", array1)
print("Array 2:", array2)
#Assign Values of Array 1 and 2 into variables X and Y respectively..
X=array1
Y=array2
#calculate covariance "Covariance = mean((X - mean(X)) Ã— (Y - mean(Y)))"
MeanX = np.mean(X)
MeanY = np.mean(Y)
Covariance = np.mean((X - MeanX)*(Y - MeanY))
print("Covariance:",Covariance)
#Calculate Standard Deviation for both arrays
StdX = np.std(X)
StdY = np.std(Y)
#Find Correlation without using corrcoef() function
Correlation = Covariance/(StdX*StdY)
print("Correlation:",Correlation)
#verify your answer using corrcoef()
print("Correlation using corrcoef():",np.corrcoef(X,Y))
```
