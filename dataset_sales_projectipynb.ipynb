**Finding Outliers using IQR**

```python
import pandas as pd
import numpy as np
#fetch data from a url or csv
url="https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/sales.csv"
#printing dataset in form string data type to make it easier to understands
arr=np.genfromtxt(url, delimiter=',', dtype=str)
print("String Array:")
print(arr)
print("\n")
#now in the form of float data type to perform calculations and basic stats
arr1=np.genfromtxt(url, delimiter=',', skip_header=1, usecols=range(1,5))
print(arr1)
```

```python
import pandas as pd
import numpy as np
Sales=arr1[:,0]
Mean=np.mean(Sales)
print("Mean:")
print(Mean)
```

```python
# percentiles
#p50=median
per25 = np.percentile(Sales, 25)
per50 = np.percentile(Sales, 50)# median
per75 = np.percentile(Sales, 75)
print("25th percentile:", per25)
print("50th percentile (median):", per50)
print("75th percentile:", per75)
#find the iqr
iqr = per75 - per25
print("Interquartile Range (IQR):", iqr)
#now we must define boundaries
lower_bound = per25 - 1.5 * iqr
upper_bound = per75 + 1.5 * iqr
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
#find the outliers
outliers = Sales[(Sales < lower_bound) | (Sales > upper_bound)]
print("Outliers:",outliers)
```
